```python
import numpy as np
from microquant.statevec import StateVector
import matplotlib.pyplot as plt
```

```python
phi = StateVector(2)

phi.Hadamard()
phi.RZ([np.pi/2, np.pi/4])
phi.Hadamard()
phi.measure()
```

```python
phi = StateVector(3)

phi.Hadamard(0)
phi.CNOT([0,1])
phi.CNOT([0,2])
phi.measure()
```

```python
phi = StateVector(2)
phi.set_state(random=True)
print(phi.state.real)
phi.CNOT([0,1])
print(phi.state.real)
phi.CNOT([1,0])
print(phi.state.real)
phi.CNOT([0,1])
print(phi.state.real)
```

```python
phi = StateVector(3)
phi.set_state(random=True)
print(phi.state.real)
phi.CNOT()
print(phi.state.real)
```

```python
x = np.linspace(0, 1, 100) * 2 * np.pi
prob_0 = []
prob_1 = []
for x_ in x:
    phi = StateVector(1)
    phi.Hadamard()
    phi.RZ(x_)
    phi.Hadamard()
    
    measure = phi.measure()
    prob_0.append(measure[0])
    prob_1.append(measure[1])
prob_0 = np.array(prob_0)
prob_1 = np.array(prob_1)
```

```python
plt.figure()

plt.plot(x, prob_0, label='prob(0)')
plt.plot(x, prob_1,  label='prob(1)')
plt.plot(x, np.sin(x/2)**2, c = 'k', ls = '--', label= 'sin^2')
plt.plot(x, np.cos(x/2)**2, c = 'k', ls = '--', label= 'cos^2')
plt.xticks(np.linspace(0,2 * np.pi, 5),[0, 0.5, 1, 1.5, 2])
plt.yticks(np.linspace(0,1,5))
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure()

plt.plot(x, prob_0, label='prob(0 | x)', c = 'tab:blue')
plt.plot(x, np.cos(x/2)**2, c = 'k', ls = '--', label= 'cos^2(x/2)')
plt.xticks(np.linspace(0,2 * np.pi, 5),[0, 0.5, 1, 1.5, 2])
plt.yticks(np.linspace(0,1,5))
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure()

plt.plot(x, prob_1,  label='prob(1 | x)', c = 'tab:orange')
plt.plot(x, np.sin(x/2)**2, c = 'k', ls = '--', label= 'sin^2(x/2)')
plt.yticks(np.linspace(0,1,5))
plt.grid()
plt.legend()
plt.show()
```
